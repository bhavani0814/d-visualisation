```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([5,7,8,7,2,17,2,9,11,12,9,6])
y=np.array([99,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array(['red','green','blue','yellow','pink','orange','grey','brown','purple','beige','black','white'])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
x=np.array([3,4,5])
y=np.array([1,8,9])
colors=np.array(['red','green','blue'])
sizes=([40,50,15])

plt.scatter(x, y, c=sizes)

plt.show()
```

```python
x=np.array([5,7,8,7,2,17,2,9,11,12,9,6])
y=np.array([99,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,1,300,600,800])
plt.scatter(x,y,s=sizes,alpha=0.2,color='r')

plt.show()
```
